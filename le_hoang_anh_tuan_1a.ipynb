Task 1: Import packages and Load data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/OpenCDP_KidStore.csv'
df = pd.read_csv(path)
df.head(30)
```

Task 2: Preprocess data

> Task 2.1: Add 'Day' column



```python
df['Day'] = ''
df.head()
```

```python
df['Day'] = df['Event Time'].str[8:10]
df.head()
```

Task 3: Reporting

> Task 3.1: What was the best day for sales? How much was earned that day?



```python
df1 = df[df['Event Type'] == 'purchase']
df1.groupby('Day').sum()['Price']
```

```python
sales_value = df1.groupby('Day').sum()['Price']
sales_value.max()
```

```python
days = range(1,32)
plt.plot(days , sales_value)
plt.xticks(days)
plt.xlabel('Days')
plt.ylabel('Revenue')
#plt.title('REVENUE BY DAY')
plt.grid()
plt.show()
```

Task 3.2: What brand has the best sales?

```python
sales_value_brand = df1.groupby('Brand').sum()['Price']
sales_value_brand.head()
```

```python
sales_value_brand_10 = sales_value_brand.sort_values(ascending = False).head(10)
sales_value_brand_10
```

```python
br = []
for Brand , Price in sales_value_brand_10.items():
  br.append(Brand)
br
```

```python
plt.bar(br,sales_value_brand_10)
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Revenue')
plt.title('TOP 10 BIGGEST BRAND BY REVENUE')
plt.show()
```

Task 3.3: What time has the best sales?

```python
df1['Hours'] = ''
df1['Hours'] = df1['Event Time'].str[11:13]
df1.head()
```

```python
sales_value_hours = df1.groupby('Hours').sum()['Price']
sales_value_hours
```

```python
hr = []
for Hours , Price in sales_value_hours.items():
  hr.append(Hours)
```

```python
plt.plot(hr,sales_value_hours)
plt.grid()
plt.title('Sales in each hour')
plt.xlabel('Hour')
plt.ylabel('Revenue')
```
